```python
df=pd.read_csv('lab1data_2.csv')
df
```

```python
df.dropna()
```

```python
df['Price']=df['Price'].str.replace(',','').astype(float)
price_mean=df['Price'].mean()
price_variance=df['Price'].var()
print("mean of price is",price_mean)
print("variance of price is",price_variance)
```

```python
wed_price=df.loc[df['Day']=='Wed','Price']
wed_mean=wed_price.mean()
print("mean of prices of wednesdays is",wed_mean)
print("the mean of prices of wednesdays alone is less when compared to the population mean")
```

```python
april_price=df.loc[df['Month']=='Apr','Price']
april_mean=april_price.mean()
print("mean of prices of April is",april_mean)
print("the mean of prices of April alone is a little high when compared to the population mean")
```

```python
df['Chg%']=df['Chg%'].str.rstrip('%').astype(float)
prob_loss=df['Chg%'].apply(lambda x: 1 if x < 0 else 0).mean()
print("probability of making a loss is",prob_loss)
```

```python
wed_val=df.loc[df['Day']=='Wed']
wed_profitprob=len(wed_val[wed_val['Chg%']>0])/len(wed_val)
print("probability of making a profit on wednesday is",wed_profitprob)
```
