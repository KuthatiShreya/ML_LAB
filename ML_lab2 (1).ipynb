```python
import numpy as np
import pandas as pd
df=pd.read_csv('lab2data_1.csv')
df
```

```python
print(df.columns)
```

```python
df=df.drop("Unnamed: 31",axis=1)
df
```

```python
df.dropna()
```

```python
df=df.replace('?','')
df
```

```python
df['TSH']=pd.to_numeric(df['TSH'],errors='coerce')
TSH_mean=df['TSH'].mean()
print(TSH_mean)
```

```python
df['T3']=pd.to_numeric(df['T3'],errors='coerce')
T3_mean=df['T3'].mean()
print(T3_mean)
```

```python
df['TT4']=pd.to_numeric(df['TT4'],errors='coerce')
TT4_mean=df['TT4'].mean()
print(TT4_mean)
```

```python
df['T4U']=pd.to_numeric(df['T4U'],errors='coerce')
T4U_mean=df['T4U'].mean()
print(T4U_mean)
```

```python
df['FTI']=pd.to_numeric(df['FTI'],errors='coerce')
FTI_mean=df['FTI'].mean()
print(FTI_mean)
```

```python
df['TBG']=pd.to_numeric(df['TBG'],errors='coerce')
TBG_mean=df['TBG'].mean()
print(TBG_mean)
```

```python
df.dtypes
```

```python
df['TSH'].fillna(TSH_mean,inplace=True)
df['T3'].fillna(T3_mean,inplace=True)
df['TT4'].fillna(TT4_mean,inplace=True)
df['T4U'].fillna(T4U_mean,inplace=True)
df['FTI'].fillna(FTI_mean,inplace=True)
df['TBG'].fillna(TBG_mean,inplace=True)
```

```python
df
```

```python
mean_TSH=df['TSH'].mean()
print("mean of TSH is",mean_TSH)
mean_T3=df['T3'].mean()
print("mean of T3 is",mean_T3)
mean_TT4=df['TT4'].mean()
print("mean of TT4 is",mean_TT4)
mean_T4U=df['T4U'].mean()
print("mean of T4U is",mean_T4U)
mean_FTI=df['FTI'].mean()
print("mean of FTI is",mean_FTI)
mean_TBG=df['TBG'].mean()
print("mean of TBG is",mean_TBG)
```

```python
var_TSH=df['TSH'].var()
print("variance of TSH is",var_TSH)
var_T3=df['T3'].var()
print("variance of T3 is",var_T3)
var_TT4=df['TT4'].var()
print("variance of TT4 is",var_TT4)
var_T4U=df['T4U'].var()
print("variance of T4U is",var_T4U)
var_FTI=df['FTI'].var()
print("variance of FTI is",var_FTI)
var_TBG=df['TBG'].var()
print("variance of TBG is",var_TBG)
```
